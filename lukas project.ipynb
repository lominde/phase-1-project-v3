# Dealing With The Missing Data

```python
# Handling Missing values

df = df.fillna("unknown")
df = df.replace("UNK", "NaN")
df['Weather.Condition'] = df['Weather.Condition'].replace('NaN','Not specified')
df['Weather.Condition'] = df['Weather.Condition'].replace('unknown','Not specified')
df['Weather.Condition'] = df['Weather.Condition'].replace('Unk','Not specified')
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].replace('Unknown','unknown')
df['Air.carrier'] = df['Air.carrier'].replace('unknown','Not defined')
df['Airport.Name'] = df['Airport.Name'].replace('unknown','Not stated')
df['Airport.Code'] = df['Airport.Code'].replace('unknown','Not specified')
df
```

```python
# Filter Unnecessary Data


df = df.drop(columns=['Report.Status'])

df

```

# Data Visualization

Using the pie chart to analyze the incident and accident of the data.

```python
counts = df['Investigation.Type'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(counts, labels=counts.index, colors=['red', 'blue'], autopct='%1.2f%%')
plt.title('Airplanes Investigation Types')
plt.show()
```

From the pie chart it is clear that the accident were the most common type of investigation at 95.57%.

#### Using the Bar chart to analyze the incident and accident of the data.

To check to which purpose the aircraft served by the air flight that lead to accident and also to analyze the models of the aircraft.



```python
# Aviation Accidents based on the Purpose of flights
purpose_of_flight_10 = df['Purpose.of.flight'].value_counts().nlargest(10).index
purpose_of_flight = df[df['Purpose.of.flight'].isin(purpose_of_flight_10)]

plt.figure(figsize=(10, 6))
sns.countplot(data=purpose_of_flight, x='Purpose.of.flight', order=purpose_of_flight_10, color = 'red')
plt.title('Top 10 Purpose of flight with the highest Number of Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

* From the above graph it is clear that Personal, Instructional, and Unknown are most prone to accidents.Luke company will need to restrict offering airplanes for this reasons and go for low use of flights with minimal accidents.

```python
# Accidents based on the make of the Airplanes
top_15_makes = df['Make'].value_counts().nlargest(15).index

top_15_df = df[df['Make'].isin(top_15_makes)]

plt.figure(figsize=(10, 6))
sns.countplot(data=top_15_df, x='Make', order=top_15_makes, color = 'blue')
plt.title('Top 15 Airplane Makes and the Number of Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

* From the makes above, Cessna, Piper, and Beech type of airplanes have the most number of accidents. Thus, the company will need to cross-check against the makes to avoid the airplanes.

```python
# Get the top 10 airplane Model by accident count
top_15_models = df['Model'].value_counts().nlargest(15).index

top_15_df = df[df['Model'].isin(top_15_models)]

plt.figure(figsize=(10, 6))
sns.countplot(data=top_15_df, x='Model', order=top_15_models, color = 'blue')
plt.title('Top 15 Airplane Model and the Number of Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

* The Luke company needs to avoid the air crafts with the most accidents numbers on the above chart.

```python
#Accidents by weather condition

weather_conditions = df['Weather.Condition'].value_counts()

# Plot as a bar chart
plt.figure(figsize=(10, 5))
weather_conditions.plot(kind='bar', title='Accidents by Weather Condition', color='orange')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Optional: Plot as a pie chart
weather_conditions.plot(kind='pie', autopct='%1.1f%%', title='Accidents by Weather Condition', figsize=(8, 8))
plt.ylabel('')
plt.show()
```

* From the Weather Condition plots above, it is notable that Pilots cause accidents under clear conditions when the visibility is very clear.
